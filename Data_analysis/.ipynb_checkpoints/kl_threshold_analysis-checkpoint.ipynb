```python
import os
import json
import numpy as np
import glob
from datetime import datetime
from collections import defaultdict


EPSILON = 1e-6  # 避免0概率
NORMAL_DIR = "baselineshort"
ATTACK_DIR = "attack"
MIX_DIR = "mix"
HTTP_DIR = "slowhttp"
SHORT_DIR = "baselineshort"
MARGIN = 0.05  # 阈值裕量

def load_all_probabilities(folder):
    result = []
    filenames = []
    for path in sorted(glob.glob(os.path.join(folder, "*.json"))):
        with open(path) as f:
            data = json.load(f)
        result.append(data["probabilities"])
        filenames.append(os.path.basename(path))
    return filenames, result

def to_full_vector(prob_dict, all_keys):
    vec = np.array([prob_dict.get(k, 0.0) for k in all_keys])
    vec = np.clip(vec, EPSILON, 1.0)
    return vec / vec.sum()

def kl_divergence(p, q):
    return np.sum(p * np.log(p / q))

# 加载正常和攻击样本
normal_names, normal_probs = load_all_probabilities(NORMAL_DIR)

# === ✅ Way 1: 逐项统计 reference（对每个状态对单独求平均） ===
prob_sums = defaultdict(float)
prob_counts = defaultdict(int)
all_keys_set = set()

for prob in normal_probs:
    for k, v in prob.items():
        prob_sums[k] += v
        prob_counts[k] += 1
        all_keys_set.add(k)

all_keys = sorted(all_keys_set)



# === ✅ Way 2: 逐项统计 多个归一化向量的平均 ===
# 获取全量状态键全集，保持排序一致性
# all_keys = set()
# for d in normal_probs + attack_probs:
#     all_keys.update(d.keys())
# all_keys = sorted(all_keys)





# 构造 reference: 每个状态转移对的平均概率
reference = {k: prob_sums[k] / prob_counts[k] for k in all_keys}
# 构建所有正常样本的向量，准备平均基准
normal_vectors = [to_full_vector(prob, all_keys) for prob in normal_probs]
# 计算平均基准概率向量 P         
P = np.mean(normal_vectors, axis=0)





# 计算所有正常样本相对于 P 的 KL 散度
normal_kls = []
print("🔵 KL divergence among normal samples (to baseline):")
for name, prob in zip(normal_names, normal_probs):
    Q = to_full_vector(prob, all_keys)
    kl = kl_divergence(P, Q)
    normal_kls.append(kl)
    print(f"  {name}: KL = {kl:.6f}")

max_normal_kl = max(normal_kls) if normal_kls else 0

```

```python
attack_names, attack_probs = load_all_probabilities(ATTACK_DIR)
# 计算所有攻击样本相对于 P 的 KL 散度
attack_kls = []
print("\n🔴 KL divergence from attacks (to baseline):")
for name, prob in zip(attack_names, attack_probs):
    Q = to_full_vector(prob, all_keys)
    kl = kl_divergence(P, Q)
    attack_kls.append(kl)
    print(f"  {name}: KL = {kl:.6f}")

min_attack_kl = min(attack_kls) if attack_kls else 0
print(f"  Min attack KL  = {min_attack_kl:.6f}")


mix_names, mix_probs = load_all_probabilities(MIX_DIR)
# 计算所有mix样本相对于 P 的 KL 散度
mix_kls = []
print("\n🔴 KL divergence from mix attacks (to baseline):")
for name, prob in zip(mix_names, mix_probs):
    Q = to_full_vector(prob, all_keys)
    kl = kl_divergence(P, Q)
    attack_kls.append(kl)
    print(f"  {name}: KL = {kl:.6f}")

min_attack_kl = min(attack_kls) if attack_kls else 0
print(f"  Min attack KL  = {min_attack_kl:.6f}")


http_names, http_probs = load_all_probabilities(HTTP_DIR)
# 计算所有mix样本相对于 P 的 KL 散度
mix_kls = []
print("\n🔴 KL divergence from attacks (to baseline):")
for name, prob in zip(http_names, http_probs):
    Q = to_full_vector(prob, all_keys)
    kl = kl_divergence(P, Q)
    attack_kls.append(kl)
    print(f"  {name}: KL = {kl:.6f}")

min_attack_kl = min(attack_kls) if attack_kls else 0
print(f"  Min attack KL  = {min_attack_kl:.6f}")


short_names, short_probs = load_all_probabilities(SHORT_DIR)
# 计算所有mix样本相对于 P 的 KL 散度
mix_kls = []
print("\n🔴 KL divergence from attacks (to baseline):")
for name, prob in zip(short_names, short_probs):
    Q = to_full_vector(prob, all_keys)
    kl = kl_divergence(P, Q)
    attack_kls.append(kl)
    print(f"  {name}: KL = {kl:.6f}")

min_attack_kl = min(attack_kls) if attack_kls else 0
print(f"  Min attack KL  = {min_attack_kl:.6f}")
```

```python
# 方法1：max_normal + margin
margin_threshold = max_normal_kl + MARGIN

# 方法2：mean + 3*std
mean_kl = np.mean(normal_kls)
std_kl = np.std(normal_kls)
stat_threshold = mean_kl + 3 * std_kl


print("\n📊 KL")
print(f"  Max normal KL  = {max_normal_kl:.6f}")
print(f"  Min attack KL  = {min_attack_kl:.6f}")
print("\n📊 Threshold Comparison:(Based on Mean + 3σ):")
print(f"  Method 1 - Max Normal KL + Margin = {margin_threshold:.6f}")
print(f"  Method 2 - Mean + 3σ              = {stat_threshold:.6f}")
# print(f"  ✅ Recommended threshold: {recommended_threshold:.6f}")






# 保存基准信息
baseline_data = {
    "probabilities": reference,
    "max_normal_kl": max_normal_kl,
    "mean_normal_kl": mean_kl,
    "std_normal_kl": std_kl,
    "min_attack_kl": min_attack_kl,
    "threshold_max_margin": margin_threshold,
    "threshold_mean_3sigma": stat_threshold,
    "used_threshold": margin_threshold,  # ← 你选哪个就填哪个
    "used_threshold2": stat_threshold,  # ← 你选哪个就填哪个
    "epsilon": EPSILON,
    "method": "KL divergence with 3σ rule",
    "description": "Statistical baseline profile for TCP state transition"
}

filename = f"baseline_profile_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
with open(filename, "w") as f:
    json.dump(baseline_data, f, indent=4)

print(f"\nBaseline profile saved to {filename}")
```
